```python
import tensorflow as tf
from tensorflow import keras

model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
model.compile(optimizer= 'sgd', loss='mean_squared_error')
```

```python
num_of_flats = (1,2,3,4,5,6,7)
price_of_houses = (10000,20000,30000,40000,50000,60000,70000)
```

```python
import numpy as np

num_of_flats = np.array(num_of_flats)
price_of_houses = np.array(price_of_houses)

model.fit(num_of_flats, price_of_houses, epochs=500)
print(model.predict(np.array([10])))
```
